## CALCULATE THE MEAN 

```python
# Applying Groupby and Agg functions for MEAN calculation
ords_prods_merge.groupby('department_id').agg({'order_number' : ['mean']})
```

Comparing means with theoretical part, we can see that the number of departments has increased. Of course it affected the means which became larger.

## CREATING LOYALTY FLAGS USING TRANSFORM() AND LOC() FUNCTIONS

```python
# Applying groupby and transform functions in order to identify the maximal amount of orders for every customer
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
ords_prods_merge.head(15)
```

```python
# Creating loyalty flags by using LOC functions 
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

```python
# Checking number of entries 
ords_prods_merge['loyalty_flag'].value_counts(dropna = False)
```

```python
ords_prods_merge.head()
```

# STEP 5

```python
# Checking basic statistics for loyalty group
ords_prods_merge.groupby('loyalty_flag').agg({'prices' : ['mean', 'min', 'max']})
```

The average price for loyal customers is lower than for New and Regular customers. (by 3 and 2 USD accordingly)

# STEP 6

```python
# IDentifying the mean price of product bought by each user by using transform function
ords_prods_merge['mean_product_price'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
ords_prods_merge[['user_id', 'mean_product_price']].head(15)
```

```python
# Creating two groups of customers: Low Spenders and High Spenders 
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] >= 10, 'spending_flag'] = 'High spender'
```

```python
ords_prods_merge.loc[ords_prods_merge['mean_product_price'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Seeing number of entries
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Demonstrating the updated table
ords_prods_merge[['user_id', 'mean_product_price', 'spending_flag']].head(15)
```

# STEP 7

```python
# Applying Transform and Groupby functions to determine the median for number of days between orders for each user
ords_prods_merge['median_days_between_orders'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders']].head(30)
```

```python
# Create order frequency flags to identify non-frequent and frequent customers
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] > 20, 'order_frequency'] = 'Non-frequent customer'
```

```python
ords_prods_merge.loc[(ords_prods_merge['median_days_between_orders'] <= 20) & (ords_prods_merge['median_days_between_orders'] > 10), 'order_frequency'] = 'Regular customer'
```

```python
ords_prods_merge.loc[ords_prods_merge['median_days_between_orders'] <= 10, 'order_frequency'] = 'Frequent customer'
```

```python
# Check number of entries (value counts) 
ords_prods_merge['order_frequency'].value_counts(dropna = False)
```


Still we have rows counts that were not labelled. Possibly they were not registered when some customers made a first order.

```python
ords_prods_merge[['user_id', 'days_since_prior_order', 'median_days_between_orders', 'order_frequency']].head(30)
```

Looking at this table we can see how the customers can be redistributed in accordance with frequency.
