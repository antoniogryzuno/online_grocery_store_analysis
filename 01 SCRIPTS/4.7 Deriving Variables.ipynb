```python
# Importing the ords_prods_merge dataset 
path_to_pickle = os.path.join(path, '02 Data', 'Prepared Data', 'ords_prods_merge1.pkl')
```

```python
ords_prods_merge = pd.read_pickle(path_to_pickle)
```

```python
#working with subset where only first one million rows shown
df = ords_prods_merge [:1000000]
```

```python
df.shape
```

```python
# Define a function for price labels

def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Applying price range function
df['price_range'] = df.apply(price_label, axis=1)
```

```python
df['price_range'].value_counts(dropna = False)
```

```python
# Finding the most expensive price 

df['prices'].max()
```

```python
# using if-statements with loc function
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
df['price_range_loc'].value_counts(dropna = False)
```

```python
# Creating of Business Day Columns 
result = []

for value in ords_prods_merge["orders_day_of_week"]:
  if value == 0:
    result.append("Busiest day")
  elif value == 4:
    result.append("Least busy")
  else:
    result.append("Regularly busy")
```

```python
#Creating new groups in plural form
result_2 = []
for value in ords_prods_merge["orders_day_of_week"]:
    if value == 0 or value == 1:
        result_2.append("Busiest days")
    elif value == 4 or value == 3:
        result_2.append("Slowest days")
    else:
        result_2.append("Regularly busy")

# Create a new column in the dataframe
ords_prods_merge['busiest_days'] = result_2
```

### STEP 4

```python
# Checking value counts in "order_hour_of_day" column.

ords_prods_merge['order_hour_of_day'].value_counts()
```

To continue with labelling, We can redistribute these counts in three equal parts: “Most orders,”: 10, 11, 14, 15, 13, 12, 16, 9. "Fewest orders: 23, 6, 0, 1, 5, 2, 4, 3 and "Average orders": all remaining values. 

```python
# Creating new columns containing new labels “Most orders,” “Average orders,” and “Fewest orders.”
result_3 = []

for value in ords_prods_merge["order_hour_of_day"]:
  if value in [10, 11, 14, 15, 13, 12, 16, 9]:
    result_3.append("Most orders")
  elif value in [23, 6, 0, 1, 5, 2, 4, 3]:
    result_3.append("Fewest orders")
  else:
    result_3.append("Average orders")
```

```python
result_3
```

```python
# Creating new column "busiest_period_of_day"
ords_prods_merge['busiest_period_of_day'] = result_3
```
