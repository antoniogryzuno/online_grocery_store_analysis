```python
custs_ords_prods_1 = custs_ords_prods.drop(columns = ['first_name', 'last_name'])
```

# STEP 3 A

```python
# Checking the number of counts for every state 
custs_ords_prods_1['state'].value_counts(dropna = False)
```

```python
# Using loc-function in order create a new column and redistribute states across regions - Northeast
custs_ords_prods_1.loc[custs_ords_prods_1['state'].isin(['Maine', 'New Hampshire', 'Vermont', 'Massachusetts', 'Rhode Island', 'Connecticut', 'New York', 'Pennsylvania', 'New Jersey']),'region_loc'] = 'Northeast'
```

```python
# Midwest
custs_ords_prods_1.loc[custs_ords_prods_1['state'].isin(['Wisconsin', 'Michigan', 'Illinois', 'Indiana', 'Ohio', 'North Dakota', 'South Dakota', 'Nebraska', 'Kansas', 'Minnesota', 'Iowa', 'Missouri']),'region_loc'] = 'Midwest'
```

```python
# South
custs_ords_prods_1.loc[custs_ords_prods_1['state'].isin(['Delaware', 'Maryland', 'District of Columbia', 'Virginia', 'West Virginia', 'North Carolina', 'South Carolina', 'Georgia', 'Florida', 'Kentucky', 'Tennessee', 'Mississippi', 'Alabama', 'Oklahoma', 'Texas', 'Arkansas', 'Louisiana']),'region_loc'] = 'South'
```

```python
# West
custs_ords_prods_1.loc[custs_ords_prods_1['state'].isin(['Idaho', 'Montana','Wyoming','Nevada','Utah','Colorado','Arizona','New Mexico','Alaska','Washington','Oregon','California','Hawaii']),'region_loc'] = 'West'
```

# STEP 3 B

```python
# Identifying differences in a spending behavior by creating a crosstab
crosstab_region_spending_habits = pd.crosstab(custs_ords_prods_1['spending_flag'], custs_ords_prods_1['region_loc'], dropna = False)
```

```python
crosstab_region_spending_habits
```

Here we can see that the low spenders and high spenders are mostly located in Southern Region. Then, it is followed by WEST, MIDWEST and NORTHEAST. There is no significant difference in spending behavior within groups.
